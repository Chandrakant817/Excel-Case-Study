```python
# importing the required library
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
  
# read a csv file
df = pd.read_csv('Tips.csv')
  
# scatter plot with regression 
# line(by default)
sns.lmplot(x ='total_bill', y ='tip', data = df)
  
# Show the plot
plt.show()
```

```python
# Scatter plot without regression line.
  
# read a csv file
df = pd.read_csv('Tips.csv')
  
# scatter plot without regression
# line.
sns.lmplot(x ='total_bill', y ='tip',
           fit_reg = False, data = df)
  
# Show the plot
plt.show()
```

```python
#Scatter plot using hue attribute for colouring out points according to the sex.
  
# read a csv file
df = pd.read_csv('Tips.csv')
  
# scatter plot using hue attribute
# for colouring out points
# according to the sex
sns.lmplot(x ='total_bill', y ='tip', 
          fit_reg = False, hue = 'sex',
          data = df)
  
# Show the plot
plt.show()
```
