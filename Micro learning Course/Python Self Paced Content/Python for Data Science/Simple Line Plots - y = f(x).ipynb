# Line Plot

### On importing matplotlib.pyplot as plt , we can draw line plots with the plt.plot() command.However, plt.plot() generates the plot inside the figure in the memory.It is displayed on screen using plt.show() function.


```python
# Importing libraries
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.colors
import seaborn as sns
# To show a line
%matplotlib inline
```

```python
x = [3, 5, 7]
y = [4, 6, 8]
plt.plot(x, y)
plt.show()
#If you forget this,plt.show() a plot object will be shown as the output instead of the plot.
```
