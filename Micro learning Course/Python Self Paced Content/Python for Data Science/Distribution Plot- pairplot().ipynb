# Pair Plots
#### Inorder to plot multiple pairwise bivariate distributions in a dataset, we make use of pairplot() function, which creates matrix of axes and shows the relationship for each pair of columns in the dataframe.

```python
# Importing libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from IPython.display import Image
```

```python
framingham = pd.read_csv(r'framingham.csv')
framingham.head()
```

```python
sns.pairplot(framingham);
```
